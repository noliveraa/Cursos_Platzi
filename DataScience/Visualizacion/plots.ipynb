### Gráfico 3D

```python
# Define the function
def my_function(x, y):
    return x * y + ((1 - y) * (1 - x))

# Generate x and y values
x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)

# Create a meshgrid from x and y
X, Y = np.meshgrid(x, y)

# Calculate Z values using the function
Z = my_function(X, Y)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show the plot
plt.show()
```

## Scatter

```python
N=50
x=np.random.rand(N)
y=np.random.rand(N)

area = (30 *np.random.rand(N))**2
colors = np.random.rand(N)

plt.scatter(x,y,s=area, c = colors, marker='<', alpha=0.5)
plt.colorbar() 
# el tamaño de la burbuja esta determinada por la variable area
# el color por la variable colors 
# con marker puedo cambiar la forma del marcados por o, x, >...
# alpha le agraga transparencia a los marcadores para que se pueda ver los que estan sobre puestos

plt.show()
```
