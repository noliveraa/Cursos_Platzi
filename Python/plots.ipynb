## Leyendas, etiquetas, ttulos, tamao

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,5))

ax1.plot(x,y, 'b',label= "label")
ax1.set_title("Grafico ax1")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.legend()

ax2.plot(y,x, 'r')
ax2.set_title("Grafico ax2")
ax2.set_xlabel("y")
ax2.set_ylabel("x")
plt.show()
```

```python
plt.plot(x,y,'bD-', label='label') #b is for blue o is the marker
plt.title("Este es el titulo")
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower right')

plt.show
```

```python
# Define the function
def my_function(x, y):
    return x * y + ((1 - y) * (1 - x))

# Generate x and y values
x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)

# Create a meshgrid from x and y
X, Y = np.meshgrid(x, y)

# Calculate Z values using the function
Z = my_function(X, Y)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show the plot
plt.show()
```
